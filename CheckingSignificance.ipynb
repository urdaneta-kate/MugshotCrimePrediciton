```python
import pandas as pd
analysis = pd.read_csv("/content/drive/MyDrive/summary new UPDATED - summary (3).csv").drop_duplicates()

analysis = demographics.merge(analysis, left_on = "ID", right_on = "mugshot_id")
```

```python
crimelist = ["Accuracy", "Homicide/Murdercrimes list","Armed Robbery", "Assault and Assault adjacent", "drugs", "manslaughter", "battery", "robbery (all)", "Bank Related crimes", "Cybercrime", "organized crime", "theft"]
```

```python
analysis["All Robbery"] = analysis["robbery (all)"]

import numpy as np
import plotly.express as px

blackallrobbery = analysis[analysis["Race"] == "Black"]["All Robbery"].values

blackallrobberyprobabilities = np.random.choice(blackallrobbery, [10000, 500], replace = True).mean(axis = 1)

whiteallrobbery = analysis[analysis["Race"] == "White"]["All Robbery"].values

whiteallrobberyprobabilities = np.random.choice(whiteallrobbery, [10000, 500], replace = True).mean(axis = 1)

(blackallrobberyprobabilities < whiteallrobberyprobabilities).mean()
```
